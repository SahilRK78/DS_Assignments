```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, recall_score, f1_score
```

```python
data.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
#Encode categorical variables
label_encoders = {}
for column in data.columns:
    le = LabelEncoder()
    data[column] = le.fit_transform(data[column])
    label_encoders[column] = le
```

```python
#correlation matrix
corr_matrix = data.corr()
```

```python
#Correlation heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
target = data[['class']]
features = data.drop('class', axis=1)

x_train, x_test, y_train, y_test = train_test_split(features, target, train_size=0.75, random_state=100)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Visualization
# Pair plot
sns.pairplot(data.iloc[:, :5].join(data['class']), hue='class')
plt.show()
```

```python
#SVM implementation
svc = SVC(C=2.0, kernel='rbf', gamma=0.1)
svc.fit(x_train, y_train)
```

```python
y_pred = svc.predict(x_test)
```

```python
y_pred
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
```

```python
print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1-score: {f1}')
```

```python

print('Classification Report:\n', classification_report(y_test, y_pred))
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred))
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(range(len(y_test)), y_test, label='Actual', alpha=0.6)
plt.scatter(range(len(y_pred)), y_pred, label='Predicted', alpha=0.6)
plt.xlabel('Samples')
plt.ylabel('Class')
plt.title('Actual vs Predicted Classes')
plt.legend()
plt.show()
```
