1.Common Hyperparameters of Decision Tree Models
Max Depth (max_depth):
Effect: Limits tree depth to prevent overfitting. Deeper trees can overfit, shallow trees can underfit.

Minimum Samples Split (min_samples_split):
Effect: Minimum samples needed to split a node. Higher values reduce overfitting.

Minimum Samples Leaf (min_samples_leaf):
Effect: Minimum samples per leaf node. Higher values prevent overfitting.

Criterion (criterion):
Effect: Split quality measure (e.g., gini, entropy). Different criteria affect tree structure.

Max Features (max_features):
Effect: Number of features to consider at each split. Fewer features increase randomness, more can improve performance.

Max Leaf Nodes (max_leaf_nodes):
Effect: Limits number of leaf nodes. Controls model complexity and prevents overfitting.

2.Difference Between Label Encoding and One-hot Encoding:

Label Encoding:
Converts categories to integers.
Usage: For ordinal data (e.g., "low," "medium," "high").

Example: "red" â†’ 0, "blue" â†’ 1.

One-hot Encoding:
Converts categories to binary columns.
Usage: For nominal data (e.g., "red," "blue," "green").

Example: "red" â†’ [1, 0, 0], "blue" â†’ [0, 1, 0].
